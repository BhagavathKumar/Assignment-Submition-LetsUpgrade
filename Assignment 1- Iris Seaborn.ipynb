```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
```

## Loading the iris dataset

```python
iris = sns.load_dataset('iris')
iris.head()
```

## Distribution Plot

```python
sns.displot(data=iris, kde = True)
```

## Joint Plot

```python
sns.jointplot(x= 'sepal_length', y= 'sepal_width',data= iris)
```

```python
sns.jointplot(x= 'sepal_length', y= 'sepal_width',data= iris, kind= 'reg')
```

```python
sns.jointplot(x= 'sepal_length', y= 'sepal_width',data= iris, hue= 'species')
```

## Scatterplot

**Scatter Plot of all the variables**

```python
plt.figure(figsize =(10,5))
sns.scatterplot(data=iris)
```

**Scatter plot of two variables along with hue argument as species**

```python
sns.scatterplot(x='sepal_length', y= 'sepal_width', hue = 'species', data= iris)
plt.legend(bbox_to_anchor = (1.3,1))                   # Used to place the legends in our desired location
```

## Pair Plot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(data=iris, kind= 'reg', diag_kind='kde', hue= 'species', palette = 'rainbow')
```
