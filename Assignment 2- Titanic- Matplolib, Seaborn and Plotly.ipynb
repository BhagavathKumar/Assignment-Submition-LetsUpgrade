## Q. Visualize the Titanic dataset  using Matplotlib, Seaborn and Plotly?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import chart_studio.plotly as py
import plotly.express as px
%matplotlib inline
sns.set_style('darkgrid')
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,iplot,plot
import cufflinks as cf
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
from jupyterthemes import jtplot
jtplot.style()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

## Different plots using Matplotlib, Seaborn and Plotly (using both methods of iplot and plotly express)

## Scatter Plot

**Matplotlib**

```python
plt.scatter(x= 'age',y= 'fare',data= titanic)
plt.xlabel('Age of Passengers')
plt.ylabel('Ticket Fare')
```

**Seaborn**

```python
sns.scatterplot(x='age',y= 'fare', data= titanic)
```

```python
sns.scatterplot(x='age',y= 'fare', data= titanic,hue= 'alive',palette= 'summer')
```

**Plotly using iplot()**

```python
cf.getThemes()
```

```python
titanic.iplot(kind= 'scatter', x= 'age',y= 'fare',mode= 'markers', theme= 'polar')
```

```python
cf.colors.scales()
```

**Plotly using express**

```python
px.scatter(x= 'age', y='fare',data_frame=titanic, color= 'alive')
```

```python
px.scatter(x= 'age', y='fare',data_frame=titanic, color= 'alive', symbol= 'class')
```

## Line Plot

**Matplotlib**

```python
x= titanic['age']
y = titanic['fare']
plt.plot(x,y)
```

**Seaborn**

```python
sns.lineplot(x='age', y='fare', data=titanic)
```

```python
sns.lineplot(x='age', y='fare', data=titanic,hue='class')
```

**Plotly using iplot()**

```python
titanic.iplot(kind= 'line', x='age', y='fare')
```

**plotly using express**

```python
px.line(data_frame= titanic, x='age', y='fare', color='class')
```

## Bar Plot

**Matplotlib**

```python
plt.bar(x= 'class', data= titanic, height= 'fare')
```

**Seaborn**

```python
sns.barplot(x='class',y='fare', data=titanic)
```

```python
sns.barplot(x='class',y='fare', data=titanic, hue= 'sex')
```

**Plotly using iplot()**

```python
titanic.iplot(kind='bar', x='class',y= 'fare')
```

**Plotly using express**

```python
px.bar(x= 'class', y= 'fare', data_frame= titanic)
```

## Box Plot

**Matplotlib**

```python
x=titanic['fare']
plt.boxplot(x)
plt.show()
```

**Seaborn**

```python
sns.boxplot(x='age', data=titanic, palette= 'hot')
```

**Plotly using iplot ()**

```python
titanic.iplot(kind= 'box')
```

**Plotly using express**

```python
px.box(x= 'age', data_frame= titanic)
```

## Spread Plot

```python
titanic.iplot(kind='spread', x='age', y='fare')       # No use in creating this plot for titanic dataset
```

```python
titanic.iplot(kind='bubble', x='age',y='fare')
```
